```python
import json
```

```python
def data_load(filename):
    with open(filename,"r") as f:
        data=json.load(f)
        
    return data    
```

```python
data=data_load("store_data.json")
```

```python
print(data)
```

```python
#clean and structure the data
def clean_data(data):
    cleaned_data=[]
    unique_user=set()
    
    #clean ratings
    text_to_num={"one":1,"two":2,"three":3,"four":4,"five":5}
    
    for user in data:
        
        raw_rating=str(user["rating"]).strip().lower()
        if raw_rating in text_to_num:
            raw_rating=text_to_num[raw_rating]
            
        user["rating"] = raw_rating
        

     # handle missing value
        missing_age=user.get("age")
        if missing_age==None:
            user["age"]=None
            
            
            

            #remove duplication
            
        name=user["name"]
            
        if name not in unique_user:
            cleaned_data.append(user)
            unique_user.add(name)
        

    return cleaned_data

    
    
```

```python
data=clean_data(data)
```

```python
#get meaningful insights from data
def get_insights(data):
    
     
   #get avg of ratings
    total_rating=0
    
    for user in data:
        
        total_rating+=float(user["rating"])
        avg_rating=total_rating/len(data)

    print(f"avg_rating={total_rating/len(data)}")


    #percentage of poor ratings
    poor_ratingcount=0
    
    for user in data:
        
        if float(user["rating"])<=3:
            poor_ratingcount+=1
    poor_ratingcount=(poor_ratingcount/len(data))*100


    print(f"poor_ratingcount={poor_ratingcount}%")
    
         
```

```python
get_insights(data)
```

```python
#product recommendation fratures
def get_recommendation(data):
    
    recommendation=[]
    
    for user in data:
        current_rec={}
        current_rec["name"]=user["name"]
        
        
        if float(user["rating"])>=4:
            current_rec["brand"]="apple"
        else:
            current_rec["brand"]="samsung"

        recommendation.append(current_rec) 

    return recommendation   
               
```

```python
get_recommendation(data)
```
