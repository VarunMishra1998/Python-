###VARIABLES IN PYTHON

A variable is a name that refers to a value stored in memory. In Python, you don’t need to declare a variable's type explicitly — Python figures it out for you.

###KEY RULES FOR VARIABLES

Variable names are case-sensitive (name and Name are different).

They can contain letters, numbers, and underscores.

Must start with a letter or an underscore (_), not a number.

Avoid Python keywords like class, def, if, etc.

###NOTE

In Python, variable names (identifiers) can contain:

Letters (A–Z, a–z)

Numbers (0–9) — but not at the beginning

Underscores (_)

###LOGICAL OPERATORS IN PYTHON

| -------- | ---------------------------- | ----------------- |

| `or`     |  Returns True if at least one of the statemebnt is True | `x < 5 or y < 10`  |

```python
### EXAMPLES OF LOGICAL OPERATORS
x=10
y=20
print(x==10 and x<y and x==y) ### In "and" case if one condition is wrong then "FALSE" output will come
print(x!=10 or x<y or x==y)  ### In "or" case if one condition is correct then "TRUE" output will come
print(x!=10)
print(not x!=10) ### "not" makes false value right and vice versa
```

###IDENTITY OPERATORS IN PYTHON
Identity operators in Python are used to compare the memory locations of two objects. In other words, they check whether two variables point to the same object in memory (not just if their values are equal).

| Operator | Description                                                    |
| -------- | -------------------------------------------------------------- |
| `is`     | Returns `True` if two variables refer to the **same object**(example : X IS Y )   |
| `is not` | Returns `True` if two variables refer to **different objects** (example : X IS NOT Y)|


```python
### EXAMPLE OF IDENTITY OPERATOR IN PYTHON
x=10
y=10
print(x is y, x==y)
print(x is not y, x!=y)
```

BETWISE OPERATORS IN PYTHON


Bitwise operators are used to perform operations on the binary representations of integers.

They work bit by bit on integers and are useful in low-level programming, such as in cryptography, networking, graphics, etc.


| Operator    | Symbol | Description                                         | Example         |
| ----------- | ------ | --------------------------------------------------- | --------------- |
| AND         | `&`    | Sets each bit to 1 **if both bits are 1**           | `5 & 3` → `1`   |
| OR          | `\|`   | Sets each bit to 1 **if at least one bit is 1**     | `5 \| 3` → `7`  |
| XOR         | `^`    | Sets each bit to 1 **if only one of the bits is 1** | `5 ^ 3` → `6`   |
| NOT         | `~`    | Inverts all the bits (1 → 0, 0 → 1)                 | `~5` → `-6`     |
| Left Shift  | `<<`   | Shifts bits **left**, filling 0s on the right       | `5 << 1` → `10` |
| Right Shift | `>>`   | Shifts bits **right**, discarding bits on the right |`5 >> 1` → `2`  |
